```python
import pandas as pd
import numpy as np
```

```python
data = {'econ':np.random.randint(10000, 400001, 100), 
        'democ': np.random.randint(-10, 11, 100) }
data = pd.DataFrame(data)
data
```

```python
# data cleaning = looking at only specific range of democracies
```

```python
df= data[(data['democ'] < -5) | (data['democ'] > 5)] 
```

```python
df.reset_index(drop=True, inplace=True)
```
